```python
import pandas as pd
users = pd.read_csv('C:/Users/adeel/Downloads/users.csv')
print(len(users))
```

```python
users.head(12)
```

```python
users['length'] = users['password'].str.len()
users['length']
```

```python
users ['too short'] = users['length'] < 8
users['too short']
```

```python
print(users['too short'].sum())
```

```python
users.head(12)
```

```python
common_passwords = pd.read_csv('C:/Users/adeel/Downloads/10_million_password_list_top_10000.txt', header=None)
common_passwords = common_passwords.squeeze()
print(common_passwords)
```

```python
users['common passwords'] = users['password'].isin(common_passwords)
users['common passwords'].head(30)
```

```python
print(users['common passwords'].sum())
```

```python
users.head(12)
```

```python
words = pd.read_csv('C:/Users/adeel/Downloads/google-10000-english.txt', header=None)
words = words.squeeze()
```

```python
users['common word'] = users['password'].str.lower().isin(words)
print(users['common word'].sum())
```

```python
users.head(12)
```

```python
users['first name'] = users['user_name'].str.extract(r'(^\w+)', expand=False)
users['last name'] = users['user_name'].str.extract(r'(\w+$)', expand=False)
```

```python
users['uses names'] = ((users['password'].str.lower()==users['first name']) | (users['password'].str.lower()==users['last name']))
```

```python
print(users['uses names'].sum())
```

```python
users.head(12)
```

```python
users['too many repeats'] = users['password'].str.contains(r'(.)\1\1\1')
```

```python
users[users['too many repeats']]
```

```python
users['bad password'] = (users['too short'] | users['common passwords'] | users['common word'] | users['uses names'] | users['too many repeats'])
```

```python
print(users['bad password'].sum())
```
